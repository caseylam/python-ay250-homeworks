```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numexpr as ne
import numpy as np
import matplotlib.pyplot as plt
```

# MOA

```python
url = "http://www.massey.ac.nz/~iabond/moa/alert2022/alert.php"
response = urlopen(url)
html = response.read()
response.close()
soup = BeautifulSoup(html,"html.parser")
# soup
# soup.find_all('th')
tE = soup.find_all('td')[4::8]
Ibase = soup.find_all('td')[6::8]
tE_list = [float(ne.evaluate(item.get_text())) for item in tE]
Ibase_list = [float(ne.evaluate(item.get_text())) for item in Ibase]

plt.hist(tE_list, bins=np.logspace(0,3,20))
plt.xscale('log')
plt.show()

plt.hist(tE_list, bins=np.linspace(0,400,20))
plt.show()

plt.plot(tE_list, Ibase_list, '.', alpha=0.5)
plt.xlabel('tE')
plt.xscale('log')
plt.ylabel('Ibase')
plt.title('MOA alerts (2021)')
plt.show()

cat = soup.find_all('td')[7::8]
cat_list = [item.get_text() for item in cat]
good_idx = [i for i,x in enumerate(cat_list) if x=='microlensing']
tE_list_ml = [tE_list[i] for i in good_idx]
Ibase_list_ml = [Ibase_list[i] for i in good_idx]

plt.plot(tE_list, Ibase_list, '.', alpha=0.5, label='all')
plt.plot(tE_list_ml, Ibase_list_ml, '.', alpha=0.5, label='ml only')
plt.xlabel('tE')
plt.xscale('log')
plt.ylabel('Ibase')
plt.title('MOA alerts (2022)')
plt.legend()
```

# KMTNet
Note: 2022 and 2021 alert pages are different, so need to munge differently.

```python
url = "https://kmtnet.kasi.re.kr/~ulens/event/2022/"
response = urlopen(url)
html = response.read()
response.close()
soup = BeautifulSoup(html,"html.parser")
#soup
tE = soup.find_all('td')[7::15][1:]
Ibase = soup.find_all('td')[10::15][1:]
Isrc = soup.find_all('td')[9::15][1:]
tE_list = [float(ne.evaluate(item.get_text().replace(u'\xa0', u''))) for item in tE]
Ibase_list = [float(ne.evaluate(item.get_text().replace(u'\xa0', u''))) for item in Ibase]
Isrc_list = [float(ne.evaluate(item.get_text().replace(u'\xa0', u''))) for item in Isrc]

plt.plot(tE_list, Ibase_list, '.', alpha=0.5)
plt.xlabel('tE')
plt.xscale('log')
plt.ylabel('Ibase')
plt.title('KMTNet alerts (2022)')
plt.show()

plt.plot(tE_list, Isrc_list, '.', alpha=0.5)
plt.xlabel('tE')
plt.xscale('log')
plt.ylabel('Isrc')
plt.title('KMTNet alerts (2022)')
plt.show()
```

```python
url = "https://ogle.astrouw.edu.pl/ogle4/ews/"
response = urlopen(url)
html = response.read()
response.close()
soup = BeautifulSoup(html,"html.parser")
# soup
# soup.find_all('td')
tE = soup.find_all('td')[8::15] 
Ibase = soup.find_all('td')[13::15]
Isrc = soup.find_all('td')[14::15]
Dmag = soup.find_all('td')[11::15] 
tE_list = [float(ne.evaluate(item.contents[0].replace(u'\n', ''))) for item in tE]
Ibase_list = [float(ne.evaluate(item.contents[0].replace(u'\n', ''))) for item in Ibase]
Isrc_list = [float(ne.evaluate(item.contents[0].replace(u'\n', ''))) for item in Isrc]
Dmag_list = [float(ne.evaluate(item.contents[0].replace(u'\n', ''))) for item in Dmag]

plt.plot(tE_list, Ibase_list, '.', alpha=0.5)
plt.xlabel('tE')
plt.xscale('log')
plt.ylabel('Ibase')
plt.title('OGLE alerts (2019)')
plt.show()

plt.plot(tE_list, Isrc_list, '.', alpha=0.5)
plt.xlabel('tE')
plt.xscale('log')
plt.ylabel('Isrc')
plt.title('OGLE alerts (2019)')
plt.show()
```
